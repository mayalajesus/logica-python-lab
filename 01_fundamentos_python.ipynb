## Entrada, saída e variáveis

### Atividade 01

Leia o nome de uma pessoa e exiba uma mensagem personalizada.

```python
def exibir_nome(nome: str) -> str:
  mensagem = 'Stay Alive, ' + nome + '!'
  return mensagem
print(exibir_nome('Dominnic'))
```

### Atividade 02

Leia nome, idade e cidade, e monte uma frase com esses dados.

```python
def pessoa(nome: str, idade: int, cidade: str) -> str:
  mensagem = f'Oi, o meu nome é {nome}, eu tenho {idade} anos de idade e moro em {cidade}.'
  return mensagem
print(pessoa('Dominnic', '30', 'Salvador'))
```

### Atividade 03

Leia dois números inteiros e exiba soma, subtração, multiplicação e divisão.

```python
def calculos(num1: int, num2: int) -> str:
  soma = num1 + num2
  sub = num1 - num2
  mult = num1 * num2
  div = num1 / num2
  mensagem = f'soma: {soma}, subtração: {sub}, multiplicação: {mult}, divisão: {div}'
  return mensagem
print(calculos(2,2))
```

### Atividade 04

Leia três notas e calcule a média aritmética.

```python
def media(num1: float, num2: float) -> str:
  calculo = (num1 + num2) / 2
  mensagem = f'resultado: {calculo}'
  return mensagem
print(media(9.3, 6.8))
```

### Atividade 05

Leia um valor em metros e converta para centímetros e milímetros.

```python
def converter_metros(num: float) -> str:
  cem = num * 100
  mil = num * 1000
  mensagem = f'centímetros: {cem}\nmílimetros: {mil}'
  return mensagem
print(converter_metros(5))
```
